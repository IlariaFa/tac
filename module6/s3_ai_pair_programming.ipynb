```python
# Votre code ici
# INSTALLATION des dÃ©pendances (Ã  exÃ©cuter une seule fois)
# !pip install spacy requests
# !python -m spacy download fr_core_news_md

import spacy
import requests

# Charger le modÃ¨le franÃ§ais de spaCy
nlp = spacy.load("fr_core_news_md")

def wikidata_search(entity):
    """
    Recherche d'une entitÃ© sur Wikidata.
    Retourne l'identifiant QID si trouvÃ©, sinon None.
    """
    url = "https://www.wikidata.org/w/api.php"
    params = {
        "action": "wbsearchentities",
        "language": "fr",
        "format": "json",
        "search": entity
    }
    response = requests.get(url, params=params).json()

    if response.get("search"):
        return response["search"][0]["id"]   # QID
    return None

def extract_entities_with_wikidata(text):
    """
    Extrait les entitÃ©s nommÃ©es d'un texte
    et les lie Ã  un identifiant Wikidata.
    """
    doc = nlp(text)
    results = []

    for ent in doc.ents:
        qid = wikidata_search(ent.text)
        results.append({
            "entitÃ©": ent.text,
            "type": ent.label_,
            "wikidata_id": qid
        })

    return results

# -----------------------------
# Test du code avec un extrait de poÃ©sie dâ€™AudreÂ Lorde
# -----------------------------
texte = """
Pour celles de nous campÃ©es sur le rivage  
debout sur le pont immuable des dÃ©cisions  
cruciales et seules  
pour celles de nous qui ne pouvons pas nous abandonner  
au rÃªve fugace des choix  
qui aiment dans les portes sâ€™en allant et revenant  
dans les heures entre les aubes  
regardant Ã  lâ€™intÃ©rieur et Ã  lâ€™extÃ©rieur  
Ã  la fois lâ€™aprÃ¨s et lâ€™avant  
cherchant un maintenant qui peut engendrer  
des futurs comme du pain dans la bouche de nos enfants  
pour que leurs rÃªves ne reflÃ¨tent plus  
la mort des nÃ´tres  
et quand nous parlons nous craignons  
que nos mots soient perdus  
ou malvenus  
mais mÃªme silencieuses  
nous restons apeurÃ©es  
alors il vaut mieux parler  
et se souvenir  
que nous nâ€™Ã©tions pas sensÃ©es survivre.
"""

entites = extract_entities_with_wikidata(texte)

for e in entites:
    print(f"EntitÃ© : {e['entitÃ©']}")
    print(f"Type : {e['type']}")
    print(f"Wikidata ID : {e['wikidata_id']}\n")
```
