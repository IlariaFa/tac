## Utiliser l'api pour analyser les pdf Camille dans le dossier pd

https://restcountries.com/

```python

#  Chemin vers mon dossier de PDF
pdf_folder = "/Users/ilaria/Desktop/STIC/traitement_auto_corpus/pdf"



# Parcourt tous les fichiers du dossier
for filename in os.listdir(pdf_folder):
    if filename.endswith(".pdf"):
        # Cherche une date dans le nom du fichier
        match = re.search(r"\d{4}-\d{2}-\d{2}", filename)
        if match:
            date_str = match.group(0)
            year, month, day = date_str.split("-")
            
            print(f"\n Fichier : {filename}")
            print(f" Date : {date_str}")
            
            #  Appelle l'API "On This Day"
            url = f"https://byabbe.se/on-this-day/{int(month)}/{int(day)}/events.json"
            response = requests.get(url)

            if response.status_code == 200:
                data = response.json()
                events = data.get("events", [])
                #  Affiche les 3 premiers événements
                for event in events[:3]:
                    print(f" {event['year']}: {event['description']}")
            else:
                print(" Échec de l’appel à l’API")
```
